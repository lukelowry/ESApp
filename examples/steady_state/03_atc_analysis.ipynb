````python
# This cell is hidden in the documentation.
from esapp import PowerWorld
from esapp.components import *
from esapp.saw._helpers import create_object_string
import ast

with open('../data/case.txt', 'r') as f:
    case_path = ast.literal_eval(f.read().strip())

pw = PowerWorld(case_path)
````

````python
areas = pw[Area, ['AreaNum', 'AreaName']]
if len(areas) >= 2:
    seller = create_object_string("Area", areas.iloc[0]['AreaNum'])
    buyer = create_object_string("Area", areas.iloc[1]['AreaNum'])
    
    pw.esa.SetData("ATC_Options", ["Method"], ["IteratedLinearThenFull"])
    pw.esa.DetermineATC(seller, buyer, do_distributed=False, do_multiple_scenarios=False)
````

````python
results = pw.esa.GetATCResults(["MaxFlow", "LimitingContingency", "LimitingElement"])

if results is not None and not results.empty:
    atc_results = results.iloc[0]
    print(f"ATC Results:")
    print(f"  Maximum Transfer: {atc_results['MaxFlow']:.1f} MW")
    print(f"  Limiting Contingency: {atc_results['LimitingContingency']}")
    print(f"  Limiting Element: {atc_results['LimitingElement']}")
````
