# Power Flow Analysis

Solves the AC power flow and inspects system voltages for violations.

```python
from gridwb import *

wb = GridWorkBench("case.pwb")
```

```python
wb.func.command("SolvePowerFlow;")

# NOTE it is in examples like this we want to replace a script command
# with an actual function call, ie wb.func.solve()
# that was the whole point of making the Adapter
```

```python
V = wb.voltage()
V.shape
```

Then we can check for violations

```python
low_v = V[abs(V) < 0.95]
len(low_v)
```
