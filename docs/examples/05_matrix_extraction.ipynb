## Network Topology

The `Network` app provides graph-based representations of the system topology. Extract the incidence matrix (branches Ã— buses):

The incidence matrix represents branches as rows and buses as columns. Each row has entries of +1 and -1 indicating the sending and receiving buses of the branch:

````python
A = wb.network.incidence()
A.shape
````

Which is of the form $(n_\text{edge}, n_\text{vert})$.
