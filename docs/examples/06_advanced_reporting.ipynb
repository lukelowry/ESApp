# Advanced Reporting and Excel Integration

This example demonstrates how to use the Adapter to generate custom reports and export data directly to Microsoft Excel, leveraging PowerWorld's built-in reporting actions.

```python
from gridwb import *
import os 

wb = GridWorkBench("case.pwb")

report_path = os.path.abspath("system_health_report.csv")
cmd = f'SaveDataWithExtra("{report_path}", "CSVCOLHEADER", "Bus", ["BusNum", "BusName", "BusPUVolt", "BusAngle"], [], "", [], ["Report_Generated_By"], ["ESA++"]);'
wb.func.command(cmd)
```

### Excel Export

```python
wb.io.esa.SendToExcel(
    "Branch", 
    "",
    ["BusNum", "BusNum:1", "LineCircuit", "LineMVA", "LineLimit", "LinePercent"],
)
```
