# Basic Data Access

Demonstrates opening a case and retrieving component data using indexing.

```python
from esapp import GridWorkBench, Bus, Gen

wb = GridWorkBench(r"case.pwb")
```

Retrieve data for buses and generators. The indexing syntax returns a pandas DataFrame, which can be used for filtering and analysis.

```python
bus_voltages = wb[Bus, "BusPUVolt"]
bus_voltages.head()
```

For example, we can get all generator data and filter for only the online generators.

```python
gens  = wb[Gen, ["GenMW", "GenStatus"]]
online_gens = gens[gens["GenStatus"] == "Closed"]
online_gens.head()
```

Modify Data

```python
wb[Gen, "GenMW"] = 100.0
```
