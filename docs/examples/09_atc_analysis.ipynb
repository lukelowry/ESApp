````python
# This cell is hidden in the documentation.
# It performs the actual case loading for the example.
from esapp import GridWorkBench
from esapp.grid import *
from esapp.saw._helpers import create_object_string
import ast

with open('case.txt', 'r') as f:
    case_path = ast.literal_eval(f.read().strip())

wb = GridWorkBench(case_path)
````

````python
areas = wb[Area, ['AreaNum', 'AreaName']]
if len(areas) >= 2:
    seller = create_object_string("Area", areas.iloc[0]['AreaNum'])
    buyer = create_object_string("Area", areas.iloc[1]['AreaNum'])
    
    wb.esa.SetData("ATC_Options", ["Method"], ["IteratedLinearThenFull"])
    wb.esa.DetermineATC(seller, buyer, do_distributed=False, do_multiple_scenarios=False)
````

````python
results = wb.esa.GetATCResults(["MaxFlow", "LimitingContingency", "LimitingElement"])

if results is not None and not results.empty:
    atc_results = results.iloc[0]
    atc_val = atc_results['MaxFlow']
    limit_ctg = atc_results['LimitingContingency']
    limit_el = atc_results['LimitingElement']
    atc_results
````
