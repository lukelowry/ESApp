# Transient Stability and Critical Clearing Time

Automating the calculation of Critical Clearing Time (CCT) for a fault and generating transient response plots to visualize system stability.

```python
from esapp import GridWorkBench, Branch

wb = GridWorkBench(r"case.pwb")
```

Initial transient stability and Calculate Critical Clearing Time (CCT) for a branch fault

Initialize the transient stability tool. Then, select a branch to apply a fault and calculate the Critical Clearing Time (CCT).

```python
wb.esa.TSInitialize()

branches = wb[Branch]
b = branches.iloc[0]
tobus = b['BusNum']
frombus = b['BusNum:1']
circuit = b['LineCircuit']

branch_str = f"[BRANCH {tobus} {frombus} {circuit}]"

wb.esa.TSCalculateCriticalClearTime(branch_str)

wb[Branch,:]
```

Generate and save plots

```python
wb.esa.TSAutoSavePlots(
    plot_names=["Generator Frequencies", "Bus Voltages"],
    ctg_names=["Fault_at_Bus_1"]
)
```
