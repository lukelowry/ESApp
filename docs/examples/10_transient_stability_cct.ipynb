````python
# This cell is hidden in the documentation.
# It performs the actual case loading for the example.
from esapp import GridWorkBench
from esapp.grid import *
from esapp.saw._helpers import create_object_string
import ast

with open('case.txt', 'r') as f:
    case_path = ast.literal_eval(f.read().strip())

wb = GridWorkBench(case_path)
````

## Critical Clearing Time Calculation

Initialize the transient stability module and calculate the critical clearing time (CCT) for a fault on a selected branch. The CCT is the maximum fault duration for which the system remains stable after fault clearing:

Initialize the transient stability tool and select the first branch for fault analysis:

````python
wb.esa.TSInitialize()

branches = wb[Branch]
b = branches.iloc[0]
tobus = b['BusNum']
frombus = b['BusNum:1']
circuit = b['LineCircuit']

branch_str = create_object_string("Branch", tobus, frombus, circuit)

wb.esa.TSCalculateCriticalClearTime(branch_str)

branches.head()
````

## Visualization

Save transient stability plots showing generator frequencies and bus voltages during the fault period:

````python
wb.esa.TSAutoSavePlots(
    plot_names=["Generator Frequencies", "Bus Voltages"],
    ctg_names=["Fault_at_Bus_1"]
)
````
