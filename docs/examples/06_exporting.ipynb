# Exporting Data

This example demonstrates how to use functions to generate custom reports and export data directly to Microsoft Excel, leveraging PowerWorld's built-in reporting actions.

```python
from gridwb import *
import os 

wb = GridWorkBench(r"case.pwb")

report_path = os.path.abspath("system_health_report.csv")
cmd = f'SaveDataWithExtra("{report_path}", "CSVCOLHEADER", "Bus", ["BusNum", "BusName", "BusPUVolt", "BusAngle"], [], "", [], ["Report_Generated_By"], ["ESA++"]);'
wb.command(cmd)
wb.esa.SaveDataWithExtra()
```

We can also export data to exel

```python
wb.esa.SendToExcel(
    "Branch", 
    "",
    ["BusNum", "BusNum:1", "LineCircuit", "LineMVA", "LineLimit", "LinePercent"],
)
```
